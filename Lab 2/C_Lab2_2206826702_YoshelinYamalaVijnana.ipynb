# Exploratory Data Analysis

**Exploratory Data Analysis (EDA)** adalah proses untuk mengeksplorasi dan memahami pola serta karakteristik utama dari suatu dataset. EDA merupakan tahap awal yang krusial dalam analisis data karena hasil dari EDA dapat mempengaruhi keputusan terkait data processing, data modeling, dan langkah-langkah analitis selanjutnya.

Proses EDA melibatkan dua komponen utama: **statistik deskriptif** dan **visualisasi data**. Melalui statistik deskriptif, kita menghitung metrik seperti mean, median, mode, dan standar deviasi untuk memahami distribusi dan variasi data. Sedangkan, melalui visualisasi data, kita dapat melihat pola dan tren dalam data menggunakan grafik seperti histogram, box plot, dan scatter plot. EDA juga dapat dibagi menjadi beberapa jenis analisis seperti **Univariate Analysis**, **Bivariate Analysis**, dan **Multivariate Analysis**.

EDA membantu kita mengidentifikasi masalah potensial dan mengarahkan analisis lebih lanjut untuk mendapatkan wawasan berharga dari data.

## Descriptive Statistics
Terdapat berbagai pengukuran yang dapat diterapkan untuk statistik deskriptif. Pengukuran ini sudah pernah Anda pelajari pada kuliah Statistika dan Probabilitas:
1. *Central tendency* (mean, median, mode)
2. *Variability* atau *dispersion* (variance, standard deviation)
3. Minimum, maximum
4. Range
5. Quartile
6. Percentile

**Central tendency data dapat digunakan untuk memahami distribusi data.** `Mean` biasanya lebih tepat digunakan untuk data dengan distribusi normal (di mana data tersebar secara simetris di sekitar pusat). Di sisi lain, `median` lebih sesuai untuk distribusi yang miring (*skewed*), di mana data condong ke satu sisi. Untuk data kategorikal, hanya `mode` yang dapat digunakan.

Namun, *central tendency* saja tidak cukup untuk menggambarkan distribusi data secara keseluruhan, karena tidak memberikan informasi mengenai tingkat penyebaran data. **Measure of variability** adalah alat ukur yang digunakan untuk memahami **variasi atau seberapa luas data tersebar**. Beberapa cara untuk mengukur variability data termasuk menggunakan `range` (jangkauan) dan `standard deviation` (simpangan baku).

Berikut adalah penggunaan statistik deskriptif (1), (2), (3) dari fitur `gross income` yang menyatakan pendapatan kotor dari suatu transaksi:

```python
gross_income = df["gross income"]
stats = [
    ['Mean', np.mean(gross_income)],
    ['Median', np.median(gross_income)],
    ['Mode', scp.mode(gross_income)[0]],
    ['Variance', np.var(gross_income)],
    ['Standard Deviation', np.std(gross_income)],
    ['Minimum', np.min(gross_income)],
    ['Maximum', np.max(gross_income)],
]
pd.DataFrame(stats, columns=['Measure', 'Value'])
```

Pada hasil di atas, bisa kita lihat bahwa measurement central tendency terbesar dimiliki oleh **mean** dengan urutan mean > median > mode.

Data tersebar sebesar 11.702970 dari nilai mean, menunjukkan bahwa ada variasi yang signifikan dalam data. Nilai maksimum adalah 49.650000 dan nilai minimum adalah 0.508500. Ini menunjukkan bahwa data cenderung tersebar di berbagai rentang nilai dengan nilai yang lebih kecil cenderung lebih banyak karena posisi median yang jauh lebih dekat dengan nilai minimum dibandingkan dengan nilai maximum. Dengan kata lain, kemungkinan data ini padat di nilai yang rendah atau padat di kiri, mengindikasikan distribusi data **skewed to the right** atau **positively skewed**.  

Untuk membuktikan pertanyaan di atas, kita akan coba visualisasikan fitur `gross income` dengan Kernel Density Estimation (KDE) plot karena dapat memberikan representasi visual tentang bagaimana data tersebut terdistribusi. Dengan KDE plot kita dapat memperkirakan distribusi data, misalnya apakah data terdistribusi secara normal, skewed, bimodal, dan lain sebagainya.

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=gross_income, shade=True)
plt.axvline(x=np.mean(gross_income), color='blue', linestyle='--', label='Mean')
plt.axvline(x=np.median(gross_income), color='red', linestyle='--', label='Median')
plt.axvline(x=scp.mode(gross_income)[0], color='brown', linestyle='--', label='Mode')
plt.xlabel('Gross Income')
plt.ylabel('Density')
plt.title('Distribusi Gross Income Transaksi')
plt.legend(fontsize=12)
plt.show()
```

Dari visualisasi distribusi dengan KDE plot di atas, dapat dengan mudah dilihat bahwa distribusi fitur `gross income` bersifat **skewed to the right** atau **positively skewed** karena longer tail ada di bagian kanan. Hal ini sesuai dengan dugaan awal kita saat menganalisis descriptive statistics. Dengan demikian, **median** adalah measurement yang lebih baik untuk merepresentasikan central tendency pada fitur `gross income`.
  
Dapat terlihat pula pada visualisasi data di atas bahwa mean, median, dan mode tidak sama, yang merupakan karakteristik umum dari distribusi skewed. Dalam kasus ini:
- Mean (ditandai dengan garis biru) lebih besar dari median (ditandai dengan garis merah) dan mode (ditandai dengan garis coklat) yang mengindikasikan bahwa beberapa nilai yang besar menarik rata-rata ke arah kanan.
- Median terletak di antara mean dan mode menjadi lebih representatif tentang central tendency dari data dalam kasus distribusi skewed ini.

Pada distribusi yang skewed, median sering kali menjadi pengukuran yang lebih baik untuk menggambarkan central tendency karena tidak terpengaruh oleh nilai ekstrem (berbeda dengan mean yang sensitif terhadap outliers). Oleh karena itu, **median** memberikan pandangan yang lebih stabil terhadap data yang terdistribusi skewed.

Kita juga bisa melihat skewness dengan method `.skew()`.

```python
# Melihat skewness dari fitur 'gross income'
gross_income.skew()
```

Untuk membandingkan distribusi data pada ketiga supermarket yang berbeda, salah satu pendekatan yang dapat kita gunakan adalah dengan visualisasi **overlapping KDE plot**. Misalnya, jika kita ingin membandingkan distribusi `gross income` pada transaksi-transaksi di ketiga supermarket, kita dapat menggunakan pendekatan overlapping KDE plot. Berikut merupakan contoh penerapannya:

```python
# Overlapping KDE plot
branch_A = df[df['Branch'] == 'A']
branch_B = df[df['Branch'] == 'B']
branch_C = df[df['Branch'] == 'C']

plt.figure(figsize=(12,10))

# Plot ketiga KDE
sns.kdeplot(branch_A['gross income'], shade=True)
sns.kdeplot(branch_B['gross income'], shade=True)
sns.kdeplot(branch_C['gross income'], shade=True)

# Plot garis yang merepresentasikan mean
plt.axvline(x=branch_A['gross income'].mean(), color='blue', linestyle='--')
plt.axvline(x=branch_B['gross income'].mean(), color='orange', linestyle='--')
plt.axvline(x=branch_C['gross income'].mean(), color='green', linestyle='--')

# Legend
plt.legend(['Branch A', 'Branch B', 'Branch C', 'Branch A Gross Income Average', 'Branch B Gross Income Average', 'Branch C Gross Income Average'], fontsize=12)

plt.show()
```

## Korelasi

Korelasi dalam data science mengacu pada hubungan statistik antara dua atau lebih fitur. Dengan memanfaatkan korelasi, kita dapat memahami bagaimana perubahan pada satu fitur terkait dengan perubahan pada fitur lainnya. Korelasi mengukur sejauh mana dua fitur bergerak bersama-sama atau memiliki ketergantungan dalam pola tertentu.

Ada dua jenis korelasi yang sering digunakan:

- Korelasi positif: Ketika dua fitur memiliki korelasi positif, ini berarti ketika satu fitur meningkat, fitur lainnya juga cenderung meningkat, dan sebaliknya. Hal ini menunjukkan adanya hubungan searah antara kedua fitur tersebut.

- Korelasi negatif: Jika dua fitur memiliki korelasi negatif, artinya ketika satu fitur naik, fitur lainnya cenderung turun, dan sebaliknya. Ini menunjukkan adanya hubungan terbalik antara kedua fitur.

Korelasi dapat diukur menggunakan koefisien korelasi, dengan **koefisien korelasi Pearson** sebagai salah satu yang paling umum digunakan. Koefisien ini memiliki rentang nilai antara -1 hingga 1:

- Nilai -1 menunjukkan korelasi negatif sempurna, artinya kedua fitur bergerak berlawanan arah secara linier.
- Nilai antara -1 dan 0: menunjukkan korelasi negatif, kedua fitur cenderung bergerak berlawanan arah secara linier.
- Nilai 0 menunjukkan tidak ada korelasi linier antara kedua fitur tersebut.
- Nilai antara 0 dan 1: menunjukkan korelasi positif, kedua fitur cenderung bergerak searah secara linier.
- Nilai +1 menunjukkan korelasi positif sempurna, artinya kedua fitur bergerak searah secara linier.

Korelasi adalah alat penting dalam analisis data karena membantu kita mengidentifikasi hubungan antar fitur, memprediksi perilaku data atau fitur, dan membuat keputusan yang lebih baik berdasarkan pemahaman tentang bagaimana fitur-fitur tersebut saling berkaitan. Namun, perlu diingat bahwa korelasi hanya menunjukkan adanya hubungan atau pola antara fitur, bukan hubungan sebab-akibat atau kausalitas. Oleh karena itu, kehati-hatian sangat diperlukan dalam menginterpretasikan hasil korelasi.

```python
# Seleksi hanya fitur numerik untuk perhitungan korelasi Pearson (kecuali gross margin percentage karena nilainya konstan)
numeric_df = df.select_dtypes(include=[np.number]).drop(columns=['gross margin percentage'])

# Implementasi korelasi Pearson hanya untuk fitur numerik
correlation_matrix = numeric_df.corr(method='pearson')
correlation_matrix
```

Berdasarkan correlation matrix di atas dapat dilihat bahwa:
- `Tax 5%`, `Total`, `cogs`, dan `gross income` memiliki korelasi positif sempurna yanng mnunjukkan bahwa terdapat hubungan linier positif sempurna yang searah. Hal ini berarti semakin tinggi Tax 5%, berarti semakin tinggi juga Total, cogs, dan gross income, dan seterusnya. Silakan explore sendiri mengapa hal ini bisa terjadi :D
- `Rating` hanya memiliki korelasi negatif yang sangat lemah dengan semua fitur lainnya yang berarti semakin tinggi Rating, ada kecenderungan kecil bahwa semakin rendah juga Unit price, Quantity, Tax 5%, Total, cogs, dan gross income
- `cogs` dan `Quantity` memiliki korelasi positif yang kuat (0.71) yang berarti semakin tinggi kuantitas produk yang dibeli, cogs atau total harga (sebelum tax) juga cenderung semakin mahal. Hal ini masuk akal karena semakin banyak produk yang dibeli berarti pengali unit price-nya jadi semakin banyak juga.
- Dan seterusnya :D

Selain metode pearson, ada metode korelasi lain yang pernah dipelajari di mata kuliah Statistika dan Probabilitas, yaitu metode spearman rho. Perbedaan mendasar antara kedua metode tersebut adalah bahwa metode pearson dihitung berdasarkan perbedaan antara nilai sebenarnya dan rata-rata, kemudian dibagi dengan standar deviasi, sedangkan metode spearman rho dihitung berdasarkan peringkat (peringkat ordinal) dari nilai setiap fitur karena penghitungan dengan pearson menggunakan rata-rata dan varians, skala pengukuran fitur harus dipastikan dalam bentuk interval dan rasio. Sebaliknya, penghitungan spearman rho menggunakan peringkat, sehingga metode ini juga dapat mengukur korelasi untuk fitur dengan skala pengukuran ordinal.

```python
correlation_matrix = numeric_df.corr(method='spearman')
correlation_matrix
```

## Data Visualization

#### Scatter Plot untuk Melihat Korelasi dan Keterhubungan 2 Fitur

```python
# Scatterplot Unit price vs gross income
df_subset = df[["Unit price", "gross income"]]
plt.figure(figsize=(8, 6))
plt.scatter(df_subset["Unit price"], df_subset["gross income"], alpha=0.5, label="Data Points")

# Perhitungan linear regression
w = df_subset["Unit price"]
z = df_subset["gross income"]
m, b = np.polyfit(w, z, 1)  # m = slope, b = intercept

# Plot regression line
plt.plot(w, m * w + b, color='red', label=f"Regression Line (y={m:.2f}x + {b:.2f})")
plt.xlabel("Unit Price")
plt.ylabel("Gross Income")
plt.title("Unit Price VS Gross Income")
plt.legend()
plt.show()
```

- Alasan penggunaan scatter plot adalah dapat menunjukkan apakah terdapat hubungan linier antara kedua fitur. Selain itu, jika memang ada hubungan linier, kita juga dapat dengan mudah melihat arahnya (apakah searah atau berlawanan arah). Jika garis regresi ke arah kanan atas berarti hubungan liniernya searah. Sebaliknya, jika garis regresi ke arah kanan bawah berarti hubungan liniernya berlawanan arah.
- Visualisasi di atas menunjukkan adanya hubungan positif yang signifikan antara `Unit price` dan `gross income`, di mana peningkatan harga per unit produk cenderung berkorelasi dengan peningkatan gross income yang terkait dengan transaksi produk tersebut. Garis regresi linier yang dihasilkan, dengan persamaan $y = 0.28x - 0.22$ mengindikasikan bahwa untuk setiap kenaikan satu unit dalam unit price, gross income meningkat sekitar 0.28 unit. Meskipun tren linier ini tampak cukup kuat, terdapat persebaran yang cukup lebar pada transaksi yang melibatkan produk dengan unit price produk yang lebih mahal, menunjukkan bahwa meskipun unit price produk merupakan prediktor penting untuk gross income, faktor lain mungkin juga berkontribusi terhadap variasi ini. Analisis lebih lanjut diperlukan untuk memahami variasi dalam gross income yang tidak sepenuhnya dijelaskan oleh unit price.

### Bar Chart untuk Melihat Rata-Rata Total Penjualan di Setiap Branch Supermarket

```python
# Menghitung rata-rata Total price di setiap transaksi pada masing-masing branch
average_total = df.groupby('Branch')['Total'].mean().reset_index()

plt.figure(figsize=(15, 6))
sorted_data = average_total.sort_values('Total', ascending=False)
ax = sns.barplot(y='Branch', x='Total', data=sorted_data, palette='viridis')
for container in ax.containers:
    ax.bar_label(container, fmt='%.2f', label_type='edge', padding=3)
plt.title("Rata-Rata Total Price di Setiap Branch")
plt.xlabel("Total")
plt.ylabel("Branch")
plt.show()
```

```python
# Jawaban 2
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset dan rating dari csv
df = pd.read_csv("supermarket_sales.csv")
ratings = df["Rating"]
mean_rating = np.mean(ratings)
median_rating = np.median(ratings)
mode_rating = ratings.mode()[0]  # Mengambil modus pertama kalau modusnya >1

# Hasil perhitungan
print(f"Mean (Rata-rata): {mean_rating:.2f}")
print(f"Median (Median): {median_rating:.2f}")
print(f"Mode (Modus): {mode_rating:.2f}")

# Visualisasi data (saya menggunakan KDE)
plt.figure(figsize=(10, 5))
sns.kdeplot(ratings, shade=True, color="lightblue", linewidth=2)
plt.axvline(mean_rating, color='blue', linestyle='dashed', linewidth=2, label=f"Mean: {mean_rating:.2f}") # Garis vertikal untuk mean, median, modus
plt.axvline(median_rating, color='green', linestyle='dashed', linewidth=2, label=f"Median: {median_rating:.2f}")
plt.axvline(mode_rating, color='red', linestyle='dashed', linewidth=2, label=f"Mode: {mode_rating:.2f}")
plt.xlabel("Rating")
plt.ylabel("Density")
plt.title("Distribusi Rating Pelanggan")
plt.legend()
plt.show()

# Sumber referensi: ChatGPT, Claude, Google
```
